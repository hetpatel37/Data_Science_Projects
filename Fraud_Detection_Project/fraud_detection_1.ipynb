```python
# scaling of time and amount feature
from sklearn.preprocessing import RobustScaler
```

```python
scale_df = df.copy()
```

```python
scale_df['Time'] = RobustScaler().fit_transform(scale_df[['Time']])
scale_df.head()
```

```python
scale_df['Amount'] = RobustScaler().fit_transform(scale_df[['Amount']])
scale_df.head()
```

```python
scale_df = scale_df.sample(frac=1, random_state=0)
```

```python
# import required libraries
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

### model-1

```python
# split train and test set
x = np.asarray(scale_df.drop('Class', axis=1))
y = np.asarray(scale_df['Class'])
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=42)

# normalization of data
std_scaler_1 = StandardScaler()
std_scaler_1.fit(x_train)
x_train = std_scaler_1.transform(x_train)

# build logistic regression model and fit it
log_reg_1 = LogisticRegression()
log_reg_1.fit(x_train, y_train)

# accuracy of model
x_test = std_scaler_1.transform(x_test)
print('accuracy:',log_reg_1.score(x_test, y_test))
print()

# classification report
y_pred = log_reg_1.predict(x_test)
print('Classification report:')
print(classification_report(y_test, y_pred, target_names=['not fraud', 'fraud']))
```

### model-2

```python
fraud_df = scale_df.loc[scale_df['Class']==1]
not_fraud_df = scale_df.loc[scale_df['Class']==0]
```

```python
# 99.67% not fraud and 0.33% fraud
# create a sample dataset
df1 = not_fraud_df.sample(n=142000, random_state=0)
df2 = fraud_df.sample(n=473, random_state=0)
new_df = pd.concat([df1, df2])
new_df = new_df.sample(frac=1, random_state=0)

# sepearte feature and target variables
x = np.asarray(new_df.drop('Class', axis=1))
y = np.asarray(new_df['Class'])

# split data into train & test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# normalization of data
std_scaler_2 = StandardScaler()
std_scaler_2.fit(x)
x_train = std_scaler_2.transform(x_train)

# build logistic regression model and fit it
log_reg_2 = LogisticRegression()
log_reg_2.fit(x_train, y_train)

# accuracy of model
x_test = std_scaler_2.transform(x_test)
print('score:',log_reg_2.score(x_test, y_test))
print()

# classification report
y_pred = log_reg_2.predict(x_test)
print('Classification report:')
print(classification_report(y_test, y_pred, target_names=['not fraud', 'fraud']))
```
